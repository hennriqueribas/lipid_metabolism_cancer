```python
#Pega o arquivo de gens e transofrma em list
arq = 'REACTOME_METABOLISM_OF_LIPIDS.v7.5.1.txt'
gensList=[]
with open(arq) as arq_gens:
    leitor = csv.reader(arq_gens, delimiter=',')
    for linha in leitor:
        gensList.extend(linha)
#gensList[0] = gensList[0][3:]
print(gensList)
```

```python
#Busca os dados no banco de dados firebrowse e cria um dataframe
dfLista = []
errGens = []
gensAux = []
testeTamanho = []
i=0
for g in gensList:
    link=''
    link+= 'http://firebrowse.org/api/v1/Samples/mRNASeq?format=json&gene='
    link+= g
    link+= '&cohort=GBM%2CGBMLGG%2CLGG&sample_type=NT%2CTM%2CTP%2CTR&protocol=RSEM&page=1&page_size=2000&sort_by=cohort'
    #verificar as p'aginas
    try:
        json_data = requests.get(link).json()
        if len(json_data['mRNASeq'])>2000:
            print(g + ': Tem mais de uma página')
#         testeTamanho.append(len(json_data['mRNASeq']))
        dfLista.append(pd.DataFrame(json_data['mRNASeq']))
#         print(len(dfLista[-1]))
#         dfLista[i].insert(loc=0, column='genes', value=[g,g,g])
        #Caso apareça Nan nos genes, há algum gen dando erro. Só se consegue ver deixando a parte de insert como comentário
        gensAux.append(g)
    except:
        errGens.append(g)
        
    i+=1
if errGens:
    print('Erro ao importar gens!!!')
df = pd.concat(dfLista, axis=0)
df.reset_index(drop=True)
```

```python
df.to_csv('KEGG_BIOSYNTHESIS_OF_UNSATURATED_FATTY_ACIDS.csv')
```
